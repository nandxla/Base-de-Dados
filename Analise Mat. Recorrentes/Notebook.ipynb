```python
import pandas as pd
import numpy as np
```

# Importando Bases
Estou importando o relatório do Mat. não recorrentes de Janeiro de 2023 até Agosto de 2024

```python
df_mat_n_recorrente_2023 = pd.read_excel("Relatório_Mat Recorrentes Jan x Dez 2023.xlsx")
df_mat_n_recorrente_2024 = pd.read_excel("Relatório_Mat. Recorrentes Jan x Ago 2024.xlsx")
```

# Tratamento de Dados
Agora, aqui está sendo feito os tratamentos nos dados, dentre eles temos:
- Transformar dados em tipo numérico (Retirando "R$", mudando a formatação do número)
- Unir as duas bases em uma só

```python
df_mat_n_recorrente_2024["VALOR UNITARIO"] = df_mat_n_recorrente_2024["VALOR UNITARIO"] \
                                                    .replace("R\$ ", "", regex=True) \
                                                    .replace(",", ".", regex=True) \
                                                    .astype("float")

df_mat_n_recorrente_2024["VALOR TOTAL"] = df_mat_n_recorrente_2024["VALOR TOTAL"] \
                                                .replace("R\$ ", "", regex=True)  \
                                                .replace("\.", "", regex=True) \
                                                .replace(",", ".", regex=True) \
                                                .astype('float')
```

```python
df_mat_n_recorrente = pd.concat([df_mat_n_recorrente_2023, df_mat_n_recorrente_2024], axis=0)
df_mat_n_recorrente.info()
```

```python
df_mat_n_recorrente.drop(columns=[
    "PROTOCOLO",
    "PREVISÃO DE ENTREGA",
    "CNPJ UNIDADE",
    "ID do Item",
    "Código do Item"
], inplace=True)
```

```python
df_mat_n_recorrente.info()
```

```python
df_mat_n_recorrente["UNIDADE"] = df_mat_n_recorrente.apply(lambda x: x["UNIDADE"].replace(f'{x["MARCA"]} ', ''), axis=1)
df_mat_n_recorrente["UNIDADE"] = df_mat_n_recorrente["UNIDADE"].str.strip()
df_mat_n_recorrente["UNIDADE"] = df_mat_n_recorrente["UNIDADE"].replace({
    " i": "1",
    " ii": "2",
    " iii": "3",
    " -": "",
}, regex=True)
```

```python
df_mat_n_recorrente["MARCA"] = df_mat_n_recorrente["MARCA"].str.strip()
df_mat_n_recorrente["Chave 2024"] = df_mat_n_recorrente["MARCA"] + df_mat_n_recorrente["UNIDADE"] 
```

```python
df_mat_n_recorrente["Chave 2024"] = df_mat_n_recorrente["Chave 2024"].str.lower()
```
